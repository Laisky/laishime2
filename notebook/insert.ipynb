```python
import re
from datetime import datetime

import pandas as pd
import numpy as np
from pymongo import Connection
```

```python
tweets = pd.read_csv('/Users/laisky/Desktop/105351466_ce802451f330f67b5f6a56d9920ab1e3d065a89d/tweets.csv')
tweets.fillna('nan', inplace=True)
tweets.rename(columns={'tweet_id': 'id', 'timestamp': 'created_at'}, inplace=True)
```

```python
from collections import defaultdict

docus = []
re_syntax = r"[\b|\s]#(\S+)"
re_compiler = re.compile(re_syntax)

statistics = {'collection': 'tweets', 'topics_count': defaultdict(lambda: 0)}
    
for i in range(tweets.shape[0]):
    docu = tweets.ix[i].to_dict()
    topics = re_compiler.findall(docu['text'])
    docu['topics'] = [t.replace('.', '_') for t in topics]
    docu['created_at'] = datetime.strptime(docu['created_at'], '%Y-%m-%d %H:%M:%S +0000')
    
    if docu['id'] != 'nan':
            docu['id'] = int(docu['id'])
            
    if docu['retweeted_status_timestamp'] != 'nan':
        docu['retweeted_status_timestamp'] = datetime.strptime(docu['retweeted_status_timestamp'], '%Y-%m-%d %H:%M:%S +0000')
    
    if docu['in_reply_to_status_id'] != 'nan':
            docu['in_reply_to_status_id'] = int(docu['in_reply_to_status_id'])

    if docu['in_reply_to_user_id'] != 'nan':
            docu['in_reply_to_user_id'] = int(docu['in_reply_to_user_id'])

    if docu['retweeted_status_id'] != 'nan':
            docu['retweeted_status_id'] = int(docu['retweeted_status_id'])

    if docu['retweeted_status_user_id'] != 'nan':
            docu['retweeted_status_user_id'] = int(docu['retweeted_status_user_id'])
    
    if docu['topics'] != []:
        for topic in docu['topics']:
            statistics['topics_count'][topic] += 1

    for k, v in docu.items():
        if v  == 'nan':
            docu[k] = None
            continue
        
        if np.isreal(v) and not isinstance(v, datetime):
            docu[k] = int(v)
                        
    docus.append(docu)
```
